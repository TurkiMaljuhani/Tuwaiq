```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df= pd.read_csv('UsedCarsSA_Clean_EN.csv')
```

```python
df
for col in df.columns:
    print(col)
```

```python
df.drop('Make', axis=1, inplace=True)
df.drop('Type', axis=1, inplace=True)
df.drop('Negotiable', axis=1, inplace=True)
df.drop('Condition', axis=1, inplace=True)
```

```python
df
```

```python
df_lr = df[['Year','Origin', 'Color','Options',
                  'Engine_Size', 'Fuel_Type', 'Gear_Type', 'Mileage', 
                    'Region', 'Price',]]
df_lr.head()
```

```python
sns.pairplot(df_lr)
plt.show()
```

```python
df_test = df_lr
```

```python
def dummies(x,df_test):
    temp = pd.get_dummies(df_test[x], drop_first = True)
    df_test = pd.concat([df_test, temp], axis = 1)
    df_test.drop([x], axis = 1, inplace = True)
    return df_test
# Applying the function to the cars_lr

df_lr = dummies('Origin',df_lr)
df_lr = dummies('Color',df_lr)
df_lr = dummies('Region',df_lr)
```

```python
df_lr.head()
```

```python
df_lr.shape
```

```python
df_lr
```

```python
#Correlation using heatmap
plt.figure(figsize = (30, 25))
sns.heatmap(df_lr.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

From the above correlation heat map, we can conclude that :-
price and Year variables are highly positively correlated (correlation coefficient = 0.31).
price and Engine_Size variables are also highly positively correlated (correlation coefficient = 0.27).

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(df_lr, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['Unnamed: 0', 'Year', 'Options', 'Engine_Size', 'Fuel_Type','Gear_Type','Mileage','Price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train
```

```python
df_train.describe()
```

```python
plt.figure(figsize = (30, 25))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
y_train = df_train.pop('Price')
X_train = df_train
```

```python
X_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X_train.drop('Origin', axis=1, inplace=True)
X_train.drop('Color', axis=1, inplace=True)
X_train.drop('Region', axis=1, inplace=True)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
rfe = RFE(lm, 10)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train.columns[rfe.support_]
```

```python
X_train_rfe = X_train[X_train.columns[rfe.support_]]
X_train_rfe.head()
```

```python
def build_model(X,y):
    X = sm.add_constant(X) #Adding the constant
    lm = sm.OLS(y,X).fit() # fitting the model
    print(lm.summary()) # model summary
    return X
    
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
X_train_new = build_model(X_train_rfe,y_train)
```

```python
X_train_new = X_train_rfe.drop(["Gear_Type"], axis = 1)
```

```python
X_train_new = build_model(X_train_new,y_train)
```

```python
checkVIF(X_train_new)
```

```python
lm = sm.OLS(y_train,X_train_new).fit()
y_train_price = lm.predict(X_train_new)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)   
```

```python
num_vars = ['Year', 'Options', 'Engine_Size', 'Fuel_Type','Gear_Type','Mileage','Price']
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])
```

```python
y_test = df_test.pop('Price')
X_test = df_test
```

```python
X_train_new = X_train_new.drop('const',axis=1)
X_test_new = X_test[X_train_new.columns]
X_test_new = sm.add_constant(X_test_new)
```

```python
y_pred = lm.predict(X_test_new)
```

```python
from sklearn.metrics import r2_score 
r2_score(y_test, y_pred)
```

```python
#EVALUATION OF THE MODEL
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)   
```
